```python
class Atm:
    def __init__(self):
        self.pin=0
        self.balance=0
        self.menu()
        
    def menu(self):
        run=True
        while(run):
            choice=int(input('''
            1.Set Pin
            2.Change Pin
            3.Withdraw
            4.Exit'''))
            if choice==1:
                self.setpin()
            elif choice==2:
                self.changepin()
            elif choice==3:
                self.withdraw()
            elif choice==4:
                run=False
    
    def setpin(self):
        self.pin = int(input("Enter a pin: "))
        self.balance = int(input("Enter your balance: "))
        print("Pin created successfully")
    
    def changepin(self):
        oldpin = int(input("Enter old pin: "))
        
        if oldpin == self.pin:
            newpin = int(input("Enter new pin: "))
            self.pin=newpin
        else:
            print("Your have enter wrong pin")
    
    def withdraw():
        wd = int(input("Enter withdrawal amount: "))
        
        if wd+1000 >= self.balance:
            print('Inufficient amount')
        else:
            self.balance-=wd
a = Atm()
a.setpin()
```
