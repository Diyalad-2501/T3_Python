```python
f = open('sample.txt','r+') #r+ means read and write (1st read then write otherwise overlap)
# f.write('welc') #overlap of character
print(f.read())
f.write('diya')
f.close()
```

```python
# we can't create file in r+ and r
f = open('sample.txt','w+')
print(f.tell())
f.write('welcome')
print(f.tell())
f.seek(0) #0 start 1 current 3 end 
# seek(offset,position)
print(f.tell())
print(f.read())
print(f.tell())
f.close()
```

```python
f = open('sample.txt','a+')
print(f.tell())
f.write('welcome')
print(f.tell())
f.seek(0) 
print(f.tell())
print(f.read())
print(f.tell())
f.close()
```
