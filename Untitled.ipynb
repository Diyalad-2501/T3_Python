```python
f = open('mbox-short.txt')
count=0
l=[]
d={}
time=[]
t={}
ans=0
c=0
for i in f:
    if i.startswith('From '):
        l.append(i.split()[1])
        time.append(i.split()[5].split(':')[0])
        count+=1
    if i.startswith('X-DSPAM-Confidence'):
        ans+=float(i.split()[1])
        c+=1
print(count)
print(ans/c)
for i in l:
    d[i]=d.get(i,0)+1
d = sorted(d.items() , key = lambda x:x[1] ,reverse=True)
print(d[0])
for i in time:
    t[i]=t.get(i,0)+1
t = sorted(t.items() , key = lambda x:x[1] ,reverse=True)
print(t)
```
